```python
import datashader as ds
import pandas as pd
from colorcet import fire
from datashader import transfer_functions as tf
```

```python
df = pd.read_csv('../data/nyc_taxi.csv', usecols=['dropoff_x', 'dropoff_y'])
df.head()
```

```python
agg = ds.Canvas().points(df, 'dropoff_x', 'dropoff_y')
tf.set_background(tf.shade(agg, cmap=fire),"black")
```

```python
import holoviews as hv
from holoviews.element.tiles import CartoLight
from holoviews.operation.datashader import datashade
hv.extension('bokeh')

map_tiles  = CartoLight().opts(alpha=1.0, width=1040, height=760, bgcolor='white')
points     = hv.Points(df, ['dropoff_x', 'dropoff_y'])
taxi_trips = datashade(points, x_sampling=1, y_sampling=1, cmap=fire, width=1040, height=760)
map_tiles * taxi_trips
```
